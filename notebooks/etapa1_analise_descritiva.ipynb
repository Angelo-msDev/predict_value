# Etapa 1 — Coleta, Pré-processamento e Análise Descritiva

**Projeto:** Sistema de Predição de Preço de Veículos  
**Base de dados:** Car Details from Car Dekho (Kaggle)  
**Problema:** Regressão — prever o preço de venda (`selling_price`)

## 1. Coleta de Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Faça upload do arquivo CSV no Colab ou ajuste o caminho
df = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
df.head()
```

## 2. Identificação das Variáveis

| Tipo | Variáveis |
|------|-----------|
| **Alvo (Y)** | `selling_price` |
| **Numéricas (X)** | `year`, `km_driven` |
| **Categóricas (X)** | `name`, `fuel`, `seller_type`, `transmission`, `owner` |

```python
df.info()
df.describe(include='all')
```

## 3. Pré-processamento

```python
print('Valores ausentes:')
print(df.isnull().sum())

print('\nDuplicatas:', df.duplicated().sum())
df = df.drop_duplicates()

df = df.dropna()
df = df[(df['selling_price'] > 0) & (df['year'] >= 1990) & (df['km_driven'] >= 0)]

df['brand'] = df['name'].apply(lambda x: str(x).split()[0])
df = df.drop('name', axis=1)

print('\nShape após limpeza:', df.shape)
df.head()
```

## 4. Análise Descritiva

```python
plt.hist(df['selling_price'], bins=50, color='steelblue', edgecolor='black')
plt.title('Distribuição do Preço de Venda')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```

```python
df['fuel'].value_counts().plot(kind='bar', color='coral')
plt.title('Quantidade por Tipo de Combustível')
plt.xlabel('Combustível')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.scatter(df['year'], df['selling_price'], alpha=0.4)
plt.title('Ano vs Preço de Venda')
plt.xlabel('Ano')
plt.ylabel('Preço')
plt.show()
```

```python
numeric = df[['year', 'km_driven', 'selling_price']]
sns.heatmap(numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
df.groupby('brand')['selling_price'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Preço Médio por Marca (Top 10)')
plt.ylabel('Preço Médio')
plt.show()
```

## 5. Principais Descobertas

- Veículos mais recentes tendem a ter preços mais altos.
- Quilometragem elevada reduz o valor estimado.
- Marcas premium (BMW, Mercedes-Benz, Audi) apresentam preços médios superiores.
- A base possui predominância de veículos a gasolina (Petrol) e transmissão manual.
